```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import scipy
from scipy import stats
```

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp','drat', 'wt', 'qsec', 'vs', 'am','gear','carb']
cars.head()
```

```python
cars.sum(0)
```

```python
cars.sum(axis= 1, numeric_only = True)
```

```python
cars.median(numeric_only = 'True')
```

```python
cars.mean(numeric_only = 'True')
```

```python
cars.mode(numeric_only ='True')
```

```python
cars.max()
```

```python
mpg = cars.mpg
mpg.idxmax()
```

```python
cars.std(numeric_only = 'True')
```

```python
cars.var(numeric_only = 'True')
```

```python
gear = cars.gear
gear.value_counts()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6)) # Reverted to previous width 10, height 6
ax = sns.lineplot(data = cars)
ax.set_xlabel('Car Features')
ax.set_ylabel('Values')
ax.set_title('Line Plot of Car Data')
ax.legend(loc='upper right', ncol=2)
plt.tight_layout()
plt.show()
```
