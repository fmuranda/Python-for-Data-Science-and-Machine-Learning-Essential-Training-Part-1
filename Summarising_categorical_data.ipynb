```python
import numpy as np
import pandas as pd
```

```python
address = 'mtcars.csv'

cars = pd.read_csv(address)
cars.columns = cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp','drat', 'wt', 'qsec', 'vs', 'am','gear','carb']
cars.index = cars.car_names
cars.head(15)
```

```python
carb = cars.carb
carb.value_counts()
```

```python
cars_cat=cars[['cyl','vs','am','gear','carb']]
cars_cat.head()
```

```python
gears_group = cars_cat.groupby('gear')
gears_group.describe()
```

```python
cars['group'] = pd.Series(cars.gear, dtype='category')
```

```python
cars['group'].dtype
```

```python
cars['group'].value_counts()
```

```python
pd.crosstab(['am'], cars['gear'])
```
