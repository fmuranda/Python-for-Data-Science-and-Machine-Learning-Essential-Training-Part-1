Import Libraries

```python
import matplotlib.pyplot as plt
import seaborn as sns

from pandas import DataFrame
```

Data

```python
data = {'names':[' steve ', ' john', 'richard', 'sarah', 'randy', 'micheal', 'julie'],
        'age':[20, 22, 20, 21, 24, 23, 22],
        'gender':[ 'Male', 'Male', 'Male', 'Female', 'Male', 'Male', 'Female'],
        'rank':[2,1,4,5,3,7,6]}

df = DataFrame(data)
df
```

Plot Bargraph - Matplotlib

```python
plt.bar(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

Plot bargraph - seaborn

```python
plot = sns.barplot(x='names', y='age', data=df)
plot.set_title('Comparing Ages')
plt.show()
```

Plot Linegraph - Matplotlib

```python
plt.plot(df['names'], df['age'])
plt.xlabel('Names')
plt.ylabel('Age')
plt.title('Comparing Ages')
plt.show()
```

Plot Linegraph - Seaborn

```python
plot = sns.lineplot(x='names', y='age', data=df)
plot.set_title('Comparing Ages')
plt.show()
```

Plot Piechart

```python
plt.pie(df['age'], labels=df['names'])
plt.title('Comparing Ages')
plt.show()
#
```

Use Seaborn Color Palette (*Seaborn has less code but makes beautiful plots*)

```python
colors = sns.color_palette('pastel')[0:5]
plt.pie(df['age'], labels=df['names'], colors=colors)
plt.title('Comparing Ages')
plt.show()
```
