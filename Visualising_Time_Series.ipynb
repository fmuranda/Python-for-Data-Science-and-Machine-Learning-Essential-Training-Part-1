```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
import seaborn as sns
```

```python
%matplotlib inline
rcParams ['figure.figsize'] = 5,4
sns.set_style('whitegrid')
```

```python
address = 'Superstore-Sales.csv'
df = pd.read_csv(address, index_col= 'Order Date', encoding= 'cp1252', parse_dates = True)
df.head()
```

To ensure the time series plot is accurate, I will first sort the DataFrame by its `Order Date` index. Then, I will plot `Order Quantity` again.

```python
df = df.sort_index()
df['Order Quantity'].plot()
plt.title('Order Quantity Over Time (Sorted)')
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
```

```python
from matplotlib import axis
df2 = df.sample(n=100, random_state=25, axis= 0)
df2 = df2.sort_index()
plt.xlabel('Order Date')
plt.ylabel('Order Quantity')
plt.title('Superstore Sales')
df2['Order Quantity'].plot()
```
